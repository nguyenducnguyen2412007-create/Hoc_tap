### Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x

```python
def is_armstrong(x: int) -> bool:
    str_x = str(x)
    n = len(str_x)  
    armstrong_sum = sum(int(digit) ** n for digit in str_x)
    return armstrong_sum == x
if __name__ == "__main__":
    number = int(input("Nhập một số nguyên: "))
    if is_armstrong(number):
        print(f"{number} là số Armstrong.")
    else:
        print(f"{number} không phải là số Armstrong.")
```

### Viết hàm kiểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x

```python
def is_perfect_square(x: int) -> bool:
    if x < 0:
        return False
    s = int(x**0.5)
    return s * s == x
x = int(input("Nhập một số nguyên: "))
if is_perfect_square(x):
    print(f"{x} là một số chính phương.")
else:    
    print(f"{x} không phải là một số chính phương.")
```

### W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên. 

```python
def is_palindrome_math(x: int) -> bool:
    if x < 0:
        return False
        
    original_number = x
    reversed_number = 0
    while x > 0:
        digit = x % 10                                 
        reversed_number = reversed_number * 10 + digit
        x = x // 10
    return original_number == reversed_number
if __name__ == "__main__":
    n = int(input("Nhập một số nguyên: "))
    if is_palindrome_math(n):
        print(f"{n} là một số đối xứng.")   
    else:
        print(f"{n} không phải là một số đối xứng.") 
```

### W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def is_magic_number(x: int) -> bool:
    if x <= 0:
        return False
        
    while x > 9:
        digit_sum = 0
        
        while x > 0:
            digit_sum += x % 10
            x = x // 10
        x = digit_sum
    return x == 1   
n = int(input("Nhập một số nguyên dương: "))
if is_magic_number(n):  
    print(f"{n} là một số ma thuật.")   
else:
    print(f"{n} không phải là một số ma thuật.")
```

### W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'. 

```python
def detect_number_type(s: str) -> str:
    s = s.strip()
    try:
        int(s)
        return 'int'
    except ValueError:
        pass
    try:
        float (s)
        if s.lower() in ['nan', 'inf', '-inf', '+inf']:
            return 'neither'
        return 'float'
    except ValueError:
        pass
    return 'neither'
n = input("Nhập số n: ")
print(detect_number_type(n))
```

```python
def is_one_product_of_others (a,b,c):
    if a == b * c:
        return True
    elif b == a * c:
        return True
    elif c == a * b:
        return True
    else:
        return False
a,b,c = map(int, input("Nhập 3 số nguyên: ").split())
if is_one_product_of_others(a,b,c):
    print("True")
else:
    print("False") 
```

### W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.

```python
def quadratic_roots_count (a, b, c):
    Delta = b**2 - 4*a*c
    if Delta > 0:
        return 2
    elif Delta == 0:
        return 1
    else:
        return 0
a,b,c = map(float, input("Nhập hệ số a, b, c: ").split())   
print("Số nghiệm của phương trình là:", quadratic_roots_count(a, b, c)) 
```

### W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không. 

```python
def are__collinear(x1,y1,x2,y2,x3,y3):
    if (y2-y1)*(x3-x2) == (y3-y2)*(x2-x1):
        return True
    else:
        return False
x1,y1 = map(float, input("Nhập tọa độ điểm A (x1 y1): ").split())
x2,y2 = map(float, input("Nhập tọa độ điểm B (x2 y2): ").split())
x3,y3 = map(float, input("Nhập tọa độ điểm C (x3 y3): ").split())
if are__collinear(x1,y1,x2,y2,x3,y3):
    print("Ba điểm A, B, C thẳng hàng.")        
else:
    print("Ba điểm A, B, C không thẳng hàng.")    
```

### W5A25 Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3), (0,3)], p = (2,1) 

```python
def point_in_quadrilateral(quad, p):
    x, y = p
    n = len(quad)
    inside = False
    p1x, p1y = quad[0]
    for i in range(n + 1):
        p2x, p2y = quad[i % n]
        if y > min(p1y, p2y):
            if y <= max(p1y, p2y):
                if x <= max(p1x, p2x):
                    if p1y != p2y:
                        xinters = (y - p1y) * (p2x - p1x) / (p2y - p1y) + p1x
                    if p1x == p2x or x <= xinters:
                        inside = not inside
        p1x, p1y = p2x, p2y
    return inside
chuoi_quad = input("Nhập tọa độ 4 điểm của tứ giác (dạng: x1 y1, x2 y2...): ")
cac_so = list(map(float, chuoi_quad.split()))
quad = [
    (cac_so[0], cac_so[1]), 
    (cac_so[2], cac_so[3]), 
    (cac_so[4], cac_so[5]), 
    (cac_so[6], cac_so[7])  
]
chuoi_diem = input("Nhập tọa độ điểm cần kiểm tra (dạng: x y): ")
px, py = map(float, chuoi_diem.split())
point = (px, py) 
print(point_in_quadrilateral(quad, point))
```

### W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
def tinh_goc (A,B,C):
  x1, y1 = A
  x2, y2 = B
  x3, y3 = C
  vector_BA = (x1 - x2, y1 - y2)
  vector_BC = (x3 - x2, y3 - y2)
  dot_product = vector_BA[0] * vector_BC[0] + vector_BA[1] * vector_BC[1]   
  do_dai_BA = (vector_BA[0]**2 + vector_BA[1]**2) ** 0.5
  do_dai_BC = (vector_BC[0]**2 + vector_BC[1]**2) ** 0.5
  if do_dai_BA == 0 or do_dai_BC == 0:
    return None  
  cos_theta = dot_product / (do_dai_BA * do_dai_BC)
  if cos_theta < -1:
        cos_theta = -1  
  elif cos_theta > 1:
        cos_theta = 1
  import math
  theta = math.acos(cos_theta)
  return math.degrees(theta)
A = tuple(map(float, input("Nhập tọa độ điểm A (x y): ").split()))
B = tuple(map(float, input("Nhập tọa độ điểm B (x y): ").split()))
C = tuple(map(float, input("Nhập tọa độ điểm C (x y): ").split()))
print("Góc tại điểm B là:", tinh_goc(A, B, C), "độ")
```

### W5A29 Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min]. 

```python
def tim_max_min(a, b, c):
    gia_tri_max = max(a, b, c)
    gia_tri_min = min(a, b, c)
    return [gia_tri_max, gia_tri_min]
if __name__ == "__main__":
    print("--- TÌM MAX VÀ MIN CỦA 3 SỐ ---")
    while True:
        try:
            chuoi_nhap = input("Nhập 3 số a, b, c (cách nhau bởi dấu cách): ")
            a, b, c = map(float, chuoi_nhap.split())
            break 
        except ValueError:
            print("LỖI: Vui lòng nhập đúng 3 số, cách nhau bởi dấu cách.")
    ket_qua = tim_max_min(a, b, c)
    print("-" * 35)
    print(f"Ba số bạn vừa nhập là: {a}, {b}, {c}")
    print(f"KẾT QUẢ [Max, Min]: {ket_qua}")
```

### W5A30 Viết hàm tính hệ số nhị thức C(n, k). 

```python
import math
def tinh_to_hop(n, k):
    if k < 0 or k > n:
        return 0
    return math.comb(n, k)
if __name__ == "__main__":
    print("--- CHƯƠNG TRÌNH TÍNH HỆ SỐ NHỊ THỨC C(n, k) ---")   
    while True:
        try:
            chuoi_nhap = input("Nhập hai số n và k (cách nhau bởi dấu cách): ")
            n, k = map(int, chuoi_nhap.split())
            if n < 0 or k < 0:
                print("=> LỖI: n và k phải là số tự nhiên lớn hơn hoặc bằng 0!")
                continue 
            break 
        except ValueError:
            print("=> LỖI: Vui lòng nhập đúng 2 số nguyên, cách nhau bởi dấu cách!")
    ket_qua = tinh_to_hop(n, k)
    print("-" * 40)
    print(f"=> KẾT QUẢ: C({n}, {k}) = {ket_qua}")
```

### W5A33 Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa, giá km, phí chờ, phụ phí ban đêm và giảm giá. 

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == 4:
        gia_mo_cua = 11000
        gia_duoi_30 = 15000
        gia_tren_30 = 12000
    elif car_type == 7:
        gia_mo_cua = 12000
        gia_duoi_30 = 17000
        gia_tren_30 = 14000
    else:
        return "Loại xe không hợp lệ (chỉ hỗ trợ 4 hoặc 7 chỗ)"
    tien_di_lai = 0
    if km <= 0.5:
        tien_di_lai = gia_mo_cua
    elif km <= 30:
        tien_di_lai = gia_mo_cua + (km - 0.5) * gia_duoi_30
    else:
        tien_di_lai = gia_mo_cua + (29.5 * gia_duoi_30) + ((km - 30) * gia_tren_30)
    tien_cho = 0
    if wait_minutes > 5:
        tien_cho = (wait_minutes - 5) * 1000
    tong_tien = tien_di_lai + tien_cho
    if start_hour >= 22 or start_hour <= 5:
        tong_tien = tong_tien * 1.20  
    if km > 50:
        tong_tien = tong_tien * 0.90  
    return int(tong_tien)
if __name__ == "__main__":
    print("--- HỆ THỐNG TÍNH CƯỚC TAXI THÔNG MINH ---")   
    try:
        loai_xe = int(input("Nhập loại xe (4 hoặc 7): "))
        gio_di = int(input("Nhập giờ bắt đầu đi (0 - 23): "))
        quang_duong = float(input("Nhập quãng đường di chuyển (km): "))
        thoi_gian_cho = int(input("Nhập thời gian chờ (phút): "))   
        ket_qua = smart_taxi_fare(loai_xe, gio_di, quang_duong, thoi_gian_cho)  
        print("-" * 40)
        print(f"=> TỔNG TIỀN CƯỚC: {ket_qua:,} VNĐ".replace(',', '.'))
    except ValueError:
        print("=> LỖI: Vui lòng nhập đúng định dạng số!")
```

### W5A34 Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau Nếu n ≤ 0 trả về 0.

```python
def sum_1_to_n(n):
    if n <= 0:
        return 0
    tong = 0
    for x in range(1, n + 1):
        chuoi_x = str(x)
        if len(set(chuoi_x)) == len(chuoi_x):
            tong += x
    return tong
if __name__ == "__main__":
    print("--- TÍNH TỔNG CÁC SỐ CÓ CHỮ SỐ KHÁC NHAU ---")    
    while True:
        try:
            n = int(input("Nhập số nguyên n: "))
            break
        except ValueError:
            print("Vui lòng chỉ nhập số nguyên!")
    ket_qua = sum_1_to_n(n)
    print("-" * 45)
    print(f"KẾT QUẢ: Tổng cần tìm = {ket_qua}")
```
