### W4A3 [Factorial] 
### Viết chương trình tính giai thừa của một số nguyên n cho trước (n!=1×2×…×n) với 0 < n < 100

```python
n = int(input("Nhập số nguyên n: "))
if n <= 0 or n >= 100:
    print("Số nhập vào không hợp lệ.")
else:
    a = 1
    for i in range(1, n + 1):
        a *= i
    print("Giai thừa của", n, "là:", a)
```

### W4A6"[SumPrimesInRange] 
### Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn [a, b]. Mỗi dòng là một bộ test (a b)."

```python
a,b = map(int, input("Nhập hai số nguyên a và b: ").split())
if a > b:
    print("Vui lòng nhập a <= b")
else:
    sum = 0
    for i in range(a, b + 1): 
        is_prime = True
        for j in range(2, int(i**0.5) + 1):
            if i % j == 0:
                is_prime = False
                break
        else:
            sum += i
    print("Tổng các số nguyên tố trong khoảng từ", a, "đến", b, "là:", sum)
```

### W4A7 "[LargestPrimeFactor] 
### Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n." 


```python
def so_nguyen_to_lon_nhat (n):
    lon_nhat = 0
    while n % 2 == 0:
        lon_nhat = 2
        n //= 2
    for i in range(3, int(n**0.5) + 1, 2):
        while n % i == 0:
            lon_nhat = i
            n //= i
    if n > 2:
        lon_nhat = n
    return lon_nhat
n = int(input("Nhập một số nguyên dương: "))
ket_qua = so_nguyen_to_lon_nhat(n)
print(f"Ước số nguyên tố lớn nhất là: {ket_qua}")                    
```

### W4A8 [ReverseAddPalindrome] 
### Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
def reverse_add_palindrome():
    n = int(input().strip())
    steps = 0
    while str(n) != str(n)[::-1]:  
        n += int(str(n)[::-1])
        steps += 1
    print(f"{steps} {n}")
if __name__ == "__main__":
    reverse_add_palindrome()
```

### W4A9[PerfectSquaresDistinctDigits] 
### Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu cách). Nếu không có, in ra (no number).

```python
def solve():
    n = int(input().strip())
    limit = int(n ** 0.5)
    results = []
    for i in range(limit + 1):
        sq = i * i
        sq_str = str(sq)
        if len(set(sq_str)) == len(sq_str):
            results.append(sq_str)
    if results:
        print(" ".join(results))
    else:
        print("(no number)")
if __name__ == "__main__":
    solve()
```

### W4A10"[LongestCollatzUpToN] 
### Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số nhỏ nhất." 

```python
def solve():
    try:
        n = int(input().strip())
    except EOFError:
        return
    memo = {1: 1}
    def get_collatz_length(num):
        if num in memo:
            return memo[num]       
        if num % 2 == 0:
            length = 1 + get_collatz_length(num // 2)
        else:
            length = 1 + get_collatz_length(3 * num + 1)
        memo[num] = length
        return length
    max_len = 0
    best_x = 0
    for i in range(1, n + 1):
        current_len = get_collatz_length(i)
        if current_len > max_len:
            max_len = current_len
            best_x = i
    print(f"{best_x} {max_len}")
solve()
```

### W4A11[EvenDivisors] 
### Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6.

```python
n = int(input("Nhập số nguyên dương n: "))
if n % 2 != 0:
    print("Không có ước số chẵn nào")
else:
    so_uoc = 0
    for i in range(2, n + 1, 2):
        if n % i == 0:
          if i % 2 == 0:
            so_uoc += 1
print("Số ước số chẵn của n là:", so_uoc)            
```

### W4A13[ClosePairNumber] 
#### Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật: Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại. Ví dụ 220 và 284 là một cặp số thân thiết 220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55, 110. Tổng của 11 ước này cho chúng ta 284 284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng vừa đúng bằng 220 Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. Nếu có, in ra màn hình "true", ngược lại, in ra "false" 

```python
a,b = map(int, input("Nhập hai số a và b: ").split())
uoc_a = 0
uoc_b = 0
for i in range(1, a):
    if a % i == 0:
        uoc_a += i
for j in range(1, b):
    if b % j == 0:
        uoc_b += j
if uoc_a == b and uoc_b == a:
    print(f"{a} và {b} là hai số bạn bè.")
else:
    print(f"{a} và {b} không phải là hai số bạn bè.")    
```

### W4A14[GCD] 
### Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình tìm ước chung lớn nhất của hai số m và n. 

```python
a,b = map(int, input("Nhập hai số a và b: ").split())
nho = min (a,b)
ucln = 1
for i in range(1, nho+1):
    if a%i == 0 and b%i == 0:
        ucln = i
print("Ước chung lớn nhất của", a, "và", b, "là:", ucln)
```
