#### Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.

```python
import math #
a = float(input("Giá trị cạnh a là "))
b = float(input("Giá trị cạnh b là "))
c = float(input("Giá trị cạnh c là "))
d = (a+b+c)/2
if a+b>c and a+c>b and b+c>a:
    s = math.sqrt(d * (d - a) * (d - b) * (d - c))
    print(f"Diện tích tam giác là {s:.1f}")
else:
    print ("Khong phai 3 canh cua tam giac")
```

#### 9. Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case

```python
a = int(input("Nhập 1 số nguyên từ 0 đến 9: "))
if a < 0 or a > 9:
    print("Lỗi.")
elif 0 <= a <= 9:
    match a:
        case 0:
           print("Không")
        case 1:
           print("Một") 
        case 2:
           print("Hai")             
        case 3:
           print("Ba")
        case 4:
           print("Bốn")
        case 5:
           print("Năm")
        case 6:
           print("Sáu")
        case 7:
           print("Bảy")
        case 8:
           print("Tám")
        case 9:
           print("Chín")          
```
