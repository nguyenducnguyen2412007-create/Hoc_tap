### Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím

```python
h = int(input("Nhập độ cao h "))
for i in range(1, h + 1):
  print(" " * (h - i), end="")
  for j in range(1, (2 * i - 1) + 1):
    if i == h or j == 1 or j == (2 * i - 1):
            print("*", end="")
    else:
            print(" ", end="")
  print(" ")         
```

### Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n = int(input("Nhập số nguyên dương n: "))
S = 0
k = 0
while S + k + 1<n:
    k += 1
    S += k
print(f"Giá trị lớn nhất của k là: {k}")
```

### Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
n = int(input("Nhập 1 số nguyên dương: "))
if n == -1:
    print("Không hợp lệ")
else:
    lon_nhat = n
    nho_nhat = n
    while True:
        n = int(input("Nhập tiếp 1 số nguyên dương: "))
        if n == -1:
           break
        if n < nho_nhat:
           nho_nhat = n
        if n > lon_nhat:    
           lon_nhat = n
print(f"Số lớn nhất là {lon_nhat}, số nhỏ nhất là {nho_nhat}")            
```

### Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input("Nhập số nguyên dương n: "))
m = n
chan = 0
le = 0
while m > 0:
    a = m%10
    if a % 2 == 0:
       chan+=1
    else:
        le+=1
    m //= 10
print(f"Số n có {chan} chữ số chẵn và {le} chữ số lẻ")           
```

### Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
A = int(input("Nhập số nguyên dương A: "))
f1 = 1
f2 = 1
if A == 1:
    print(f"Số Fibonacci lớn nhất không vượt quá {A} là: 1")
else:
    f_next = f1 + f2         
    while f_next <= A:
        f1 = f2          
        f2 = f_next       
        f_next = f1 + f2
print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {f2}")        
```

### Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.

```python
n = int(input("Nhập số nguyên dương n: "))
m = n
nguyen_to = 0
while m > 0:
    chu_so = m % 10
    if chu_so == 2 or chu_so == 3 or chu_so == 5 or chu_so == 7:
        nguyen_to += 1
    m //= 10        
print(f"Số {n} có {nguyen_to} chữ số là số nguyên tố.")
```
