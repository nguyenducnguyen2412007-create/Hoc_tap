#### W2A1 [Hello World!]
#### Máy tính giao tiếp với con người thông qua các thiết bị thu nhận các yêu cầu của người dùng (input devices) như bàn phím, con chuột, microphone và các thiết bị hiển thị kết quả tính toán (output devices) như màn hình, loa. Trong bài này, chúng ta sẽ làm việc với thiết bị hiển thị kết quả tính toán đơn giản nhất, đó là màn hình văn bản (text screen). Màn hình văn bản được dùng để hiển thị các thông tin dưới dạng văn bản từ chữ cái (characters), chữ số (digits) đến các dòng văn bản (text line), các đoạn văn bản (text paragraph). Trong đa số trường hợp, màn hình văn bản còn được gọi là thiết bị xuất chuẩn (standard output device) bởi vì nó là thiết bị hiển thị và đầu ra mặc định của các chương trình. Trong python hàm print() được dùng để hiển thị thông tin ra màn hình (console), nó là một trong những công cụ quan trọng nhất để: Hiển thị kết quả cho người dùng, Kiểm tra, debug chương trình khi lập trình, Xuất thông tin định dạng đẹp... #### print(*objects, sep=' ', end='\n', file=sys.stdout, flush=False). 
#### Các đối số hay dùng:
#### '- *objects: Chuỗi, số, danh sách, v.v... cần in ra.

#### '- flush: Ép ghi ra màn hình ngay lập tức.
#### print() là một trong những hàm được Python cung cấp sẵn, bạn có thể dùng nó mà không cần import thư viện gì hết.Nó được định nghĩa trong mô-đun builtins, là mô-đun mặc định luôn được nạp khi Python chạy.Như vậy, toàn bộ một chương trình in ra dòng chữ "Hello World!" như sau: print("Hello World!").

### W2A3 [ModAndDiv]
### Viết chương trình nhận vào hai số nguyên a, b từ bàn phím. In ra tổng, hiệu, tích, phần nguyên, phần dư và kết quả chia thực của phép chia a cho b (lấy đến 2 chữ số phần thập phân) theo định dạng đúng của output

```python
a = int(input("Nhập số tự nhiên a: "))
b = int(input("Nhập số tự nhiên b: "))
c = a+b
d = a-b
e = a*b
f = a//b
g = a%b
h = a/b
print(f"Tổng: {c}")
print(f"Hiệu: {d}")
print(f"Tích: {e}") 
print(f"Phần nguyên: {f}")
print(f"Phần dư: {g}")
print(f"Phần thực: {h:.2f}")
```

### W2A4 [AverageOfSix]
### Viết chương trình python để tính điểm trung bình môn học của một học sinh trong lớp. Có tổng cộng 6 đầu điểm: a1, b1, c1, a2, b2, a3. Trong đó:
#### • Các điểm a1, b1, c1 được nhân hệ số 1,
### • Các điểm a2, b2 được nhân hệ số 2,
### • Điểm a3 được nhân hệ số 3.
### Công thức tính điểm trung bình là: TB=((a1+b1+c1)+(a2+b2)×2+a3×3)/10
### Đầu vào: Trong bài tập này, dữ liệu vào chỉ gồm duy nhất một dòng bao gồm 6 số thể hiện cho 6 đầu điểm a1, b1, c1, a2, b2, a3 theo đúng thứ tự đó. Trong đó, hai đầu điểm liên tiếp nhau sẽ được phân tách nhau bởi duy nhất 1 dấu cách.Đầu ra: Đầu ra là giá trị điểm trung bình môn học của 6 môn học, lấy đến 1 chữ số phần thập phân.

```python
a1,b1,c1,a2,b2,a3 = map(float,input("Nhập điểm các môn học: ").split())
c = ((a1+b1+c1)+(a2+b2)*2+a3*3)/10
print(f"Điểm trung bình môn học là: {c:.1f}")
```

### W2A6 [StringDataType]
### “Chuỗi” trong python là các kí tự được xâu lại với nhau như những hạt trên một chiếc dây chuyền. Chính vì thế “chuỗi” còn được gọi là “xâu”. Một chuỗi gồm các kí tự nằm trong dấu nháy đơn '' hoặc nháy đôi "”.• Mỗi chuỗi (str) là một chuỗi các kí tự trong bảng mã Unicode kể từ Python 3 trở đi (Unicode là một tiêu chuẩn mã hóa kí tự toàn cầu, cho phép biểu diễn hầu hết mọi kí tự trong tất cả các ngôn ngữ trên thế giới, bao gồm cả kí hiệu, biểu tượng, emoji,….). Bạn có thể dùng ord() và chr() để làm việc với mã Unicode. 
### Hãy nhập vào một kí tự chữ cái thường ‘a’…’z’, in ra mã Unicode của nó sau đó chuyển nó thành kítự hoa tương ứng, in kí tự hoa này ra màn hình.

```python
a = input("Nhập 1 chữ cái thường: ")
b = ord(a)
if b<97 or b>122:
    print("Lỗi.")
else:    
    c = chr(b-32)
    print(f"Chữ in hoa tương ứng là: {c}")
```

### W2A12 Cho một khối Rubik có độ dài cạnh là n, hãy trả về số lượng miếng dán riêng lẻ cần thiết để bao phủ toàn bộ khối lập phương. Ví dụ, khối rubic có độ dài cạnh là 1 cần 6 miếng dán riêng lẻ. (Lưu ý: một khối rubic có 6 mặt)

### W2A13 Nhập vào từ bàn phím hai số nguyên dương \a và \b, viết chương trình trả về hàng đơn vị của phép tính tích giữa \a và \b

```python
try:
    a, b = map(int, input("Nhập hai số nguyên dương a và b: ").split())
    if a > 0 and b > 0:
        c = a * b
        d = c % 10
        print(d)
    else:
        print("Nhập số nguyên dương!")
except ValueError:
    print("Lỗi định dạng!")
```
