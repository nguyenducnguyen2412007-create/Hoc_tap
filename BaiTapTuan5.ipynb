### W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 

```python
def kiem_tra_so_nguyen_to(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
n = int(input("Nhập một số nguyên: "))
if kiem_tra_so_nguyen_to(n):
    print(f"{n} là số nguyên tố.")
else:
    print(f"{n} không phải là số nguyên tố.")
```

### W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.  

```python
def kiem_tra_so_hoan_hao (n):
    if n < 2:
        return False
    s = 1
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            s += i
            if i != n // i:
                s += n // i
    return s == n
n = int(input("Nhập một số nguyên dương: "))
if kiem_tra_so_hoan_hao(n):
    print(f"{n} là số hoàn hảo.")
else:
    print(f"{n} không phải là số hoàn hảo.")
```

### W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)

```python
def tinh_giai_thua (n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * tinh_giai_thua(n - 1)
n = int(input("Nhập một số nguyên dương: "))
if n < 0:
    print("Vui lòng nhập một số nguyên dương.") 
else:
    print(f"Giai thừa của {n} là: {tinh_giai_thua(n)}")
```

#### W5A7 Viết hàm giả lập một máy tính bỏ túi. 
#### Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat \) 
#### Biết các toán tử đơn giản thực hiện phép toán lần lượt là: 
#### Phép trừ \( - \) 
#### Phép cộng \( + \) 
#### Phép chia \( / \) 
#### Phép nhân \( * \) 
#### Kết quả làm tròn đến chữ số thập phân thứ hai. 
#### Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)). 
#### Đầu ra: Kết quả phép toán.

```python
def may_tinh_bo_tui(num1, num2, operat):
    if operat == '+':
        result = num1 + num2
    elif operat == '-':
        result = num1 - num2
    elif operat == '*':
        result = num1 * num2
    elif operat == '/':
        # Kiểm tra điều kiện chia cho 0
        if num2 == 0:
            return "Lỗi: Không thể chia cho 0!"
        result = num1 / num2
    else:
        return "Lỗi: Toán tử không hợp lệ!"
    return round(result, 2)
num1,num2 = float(input("Nhập số thứ nhất: ")), float(input("Nhập số thứ hai: "))
operat = input("Nhập phép toán (+, -, *, /): ")
result = may_tinh_bo_tui(num1, num2, operat)
print(f"Kết quả: {result}")
```

#### W5A8 Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2. 
#### Cho hai số nguyên  x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này. Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn.

```python
def hamming_distance_bitwise(x: int, y: int) -> int:
    xor_val = x ^ y
    distance = 0
    while xor_val:
        distance += 1
        xor_val &= (xor_val - 1)
    return distance
x, y = map(int, input("Nhập hai số nguyên : ").split())
print(f"Khoảng cách Hamming giữa {x} và {y} là:", hamming_distance_bitwise(x, y)) 
```

### W5A9 Viết hàm trả về tổng các chữ số của số nguyên dương n 

```python
def tong_chu_so_toan_hoc(n: int) -> int:
    tong = 0
    while n > 0:
        tong += n % 10  
        n //= 10        
    return tong
n = int(input("Nhập một số nguyên dương: "))
print("Tổng các chữ số của n     là:", tong_chu_so_toan_hoc(n)) 
```

### W5A10 Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành 'z', 'b' thành 'b' và 'c' thành 'x'. Viết hàm  trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai từ đầu vào có cùng độ dài.

```python
def la_dang_cau(a, b):
    map_a_to_b = {}
    map_b_to_a = {}
    for char_a, char_b in zip(a, b):
        if char_a in map_a_to_b:
            if map_a_to_b[char_a] != char_b:
                return False
        else:
            map_a_to_b[char_a] = char_b
        if char_b in map_b_to_a:
            if map_b_to_a[char_b] != char_a:
                return False
        else:
            map_b_to_a[char_b] = char_a
    return True
a = input("Nhập chuỗi a: ")
b = input("Nhập chuỗi b: ")
if la_dang_cau(a, b):
    print("Hai chuỗi là dẳng cấu của nhau.")
else:
    print("Hai chuỗi không phải là dẳng cấu của nhau.")
```

### W5A12 Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và 0 <= |r| < |divisor|. 

```python
def divide_with_remainder(dividend, divisor):
    if divisor == 0:
        raise ZeroDivisionError("Số chia (divisor) không thể bằng 0.")
    q, r = divmod(dividend, divisor)
    return (q, r)
dividend = int(input("Nhập số bị chia (dividend): "))
divisor = int(input("Nhập số chia (divisor): "))
if divisor == 0:
    print("Lỗi: Số chia (divisor) không thể bằng 0.")
else:
    quotient, remainder = divide_with_remainder(dividend, divisor)
    print(f"Thương (quotient): {quotient}, Số dư (remainder): {remainder}")
```

### W5A14 Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op. 

```python
def compute (a, b, operator):
    if operator == '+':
        return a + b
    elif operator == '-':
        return a - b
    elif operator == '*':
        return a * b
    elif operator == '/':
        if b == 0:
            raise ValueError("Số chia không thể bằng 0.")
        return a / b
    else:
        raise ValueError("Toán tử không hợp lệ.")
a,b = map(float, input("Nhập hai số a và b: ").split())
operator = input("Nhập toán tử (+, -, *, /): ")
if operator in ['+', '-', '*', '/']:
    try:
        result = compute(a, b, operator)
        print(f"Kết quả: {result}")
    except ValueError as e:
        print(e)    
```

### W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông. 

```python
def is_right_triangle(a, b, c):
    sides = sorted([a, b, c])
    return sides[0]**2 + sides[1]**2 == sides[2]**2
a,b,c = map(float, input("Nhập độ dài ba cạnh của tam giác: ").split())
if is_right_triangle(a, b, c):
    print("True")
else:
    print("False")
```
