### W3A1
### Nhập vào một số nguyên dương có 3 chữ số. Tìm và in ra màn hình “SỐ NGUYÊN” đảo ngược của n. Không dùng kiểu chuỗi. Chú ý: kiểm tra chỉ xử lý các số nguyên dương có đúng 3 chữ số. Các số khác không xử lý.

```python
n = input("Nhập vào một số nguyên: ")
if int(n) <100 or int(n) > 999:
    print("Số bạn nhập không phải là số có 3 chữ số.")  
else:
    a = int(n) // 100
    b = (int(n) % 100) // 10
    c = int(n) % 10
    n_dao = c * 100 + b * 10 + a
    print("Số đảo ngược của", n, "là", n_dao)   
```

### Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0

### W3A11[Phân loại tam giác] 
### Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay "Tam giác thường".

```python
a,b,c = map(int, input("Nhập ba số nguyên a, b, c: ").split())
if a+b>c and a+c>b and b+c>a:
    if a==b and b==c:
        print("Tam giác đều")
    elif a==b or b==c or a==c:
        print("Tam giác cân") 
    else:
        print("Tam giác thường")        
else:    
    print("Không phải là tam giác")
```

### W3A16[Rounding] 
### Nhập vào một số thực: in ra kết quả làm tròn lên, tròn xuống, làm tròn tới số nguyên gần nhất (không được sử dụng hàm sẵn có của pyton). Các kết quả cách nhau một dấu cách


```python
n = float(input("Nhập số thực n: "))
a = int(n)+1
b = int(n)
if abs(n-a) < abs(n-b): 
    print(a,b,a)
else: 
    print(a,b,b)
```

### Gợi ý : Tìm công bội (b / a) nếu b chia hết cho a, sau đó lấy b nhân công bội và so sánh vs c, c nhân công bội và so sánh vs d.

```python
a,b,c,d = map(float, input("Nhập bốn số thực: ").split())
if b%a == 0 and c%b == 0 and d%c == 0:
    if b/a == c/b == d/c:
        print("Bốn số tạo thành một cấp số nhân.")
    else:        
        print("Bốn số không tạo thành một cấp số nhân.")
else:
    print("Bốn số không tạo thành một cấp số nhân.")            
```
