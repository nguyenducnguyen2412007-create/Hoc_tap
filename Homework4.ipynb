### W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 

```python
n = int(input("Nhập số nguyên dương n: "))
m = n
if n <= 0:
    print(f"{n} KHÔNG phải là số có dạng 2^k.")
else:
    while m % 2 == 0:
        m //= 2
    if m == 1:
           print(f"{n} là số có dạng 2^k.")
    else:
           print(f"{n} KHÔNG phải là số có dạng 2^k.")
```

### W4A21 Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input("Nhập số nguyên dương n: "))
m = n
a = 0
while m > 0:
    b = m%10
    a += b    
    m //= 10
print(f"Tổng các chữ số của n là: {a}") 
```

### W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input("Nhập số nguyên dương n: "))
m = n
chan = 0
le = 0
while m > 0:
    a = m%10
    if a % 2 == 0:
       chan+=1
    else:
        le+=1
    m //= 10
print(f"Số n có {chan} chữ số chẵn và {le} chữ số lẻ") 
```

### W4A25
### Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
n = int(input("Nhập 1 số nguyên dương: "))
if n == -1:
    print("Không hợp lệ")
else:
    lon_nhat = n
    nho_nhat = n
    while True:
        n = int(input("Nhập tiếp 1 số nguyên dương: "))
        if n == -1:
           break
        if n < nho_nhat:
           nho_nhat = n
        if n > lon_nhat:    
           lon_nhat = n
print(f"Số lớn nhất là {lon_nhat}, số nhỏ nhất là {nho_nhat}")            
```

#### W4A26 [Fibonaci1] 
### Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó 
### Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
A = int(input("Nhập số nguyên dương A: "))
f1 = 1
f2 = 1
if A == 1:
    print(f"Số Fibonacci lớn nhất không vượt quá {A} là: 1")
else:
    f_next = f1 + f2         
    while f_next <= A:
        f1 = f2          
        f2 = f_next       
        f_next = f1 + f2
print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {f2}")        
```

#### W4A27
### Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng) 
### Dùng cách: for trên index; KHÔNG dùng hàm split(), len()

```python
a = input("Nhập 1 chuỗi: ")
m = 0
n = True
for k, ki_tu in enumerate(a):
    if not ki_tu.isalpha() and ki_tu != ' ':
        print("Lỗi")
        n = False
        break
    elif ki_tu.isalpha():
        if k == 0 or a[k-1] == ' ':
            m += 1
if n == True:
    print(m)
```

### W4A28
### Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi 
### Dùng cách: for trên index; KHÔNG dùng hàm split()

```python
chuoi = input("Nhập chuỗi: ")
tu_dau_tien = ""
for i in range(len(chuoi)):
    ky_tu = chuoi[i]
    if ky_tu != ' ':  
        tu_dau_tien += ky_tu  
    else:
        if tu_dau_tien != "":
            break 
if tu_dau_tien != "":
    print("Từ đầu tiên trong chuỗi là:", tu_dau_tien)
else:
    print("Chuỗi rỗng hoặc chỉ toàn khoảng trắng!")
```

### W4A29
### Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành thường và thường thành hoa.

```python
chuoi = input("Nhập chuỗi: ")
chuoi_moi = ""
for i in range(len(chuoi)):
    ky_tu = chuoi[i]
    if ky_tu.isupper():
        chuoi_moi += ky_tu.lower()
    elif ky_tu.islower():
        chuoi_moi += ky_tu.upper()
    else:
        chuoi_moi += ky_tu
print("Chuỗi sau khi chuyển đổi là:", chuoi_moi)
```

### W4A31
### Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in thường, bao nhiêu ký tự số

```python
chuoi = input("Nhập một chuỗi: ")
in_hoa = 0
in_thuong = 0
so = 0
for ki_tu in chuoi:
    if ki_tu.isupper():
        in_hoa += 1
    elif ki_tu.islower():
        in_thuong += 1
    elif ki_tu.isdigit():
        so += 1
print(f"Số ký tự in hoa: {in_hoa}")
print(f"Số ký tự in thường: {in_thuong}")
print(f"Số ký tự số: {so}")        
```
